# Deep Learning Models


## LSTM

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
from keras.models import Sequential
from keras.layers import LSTM, Dense
plt.rcParams['font.family'] = 'Times New Roman'
# Define the model architecture
model = Sequential()
model.add(LSTM(64, input_shape=(128, 1)))  # Assuming input_shape is (timesteps, features)
model.add(Dense(1, activation='sigmoid'))  # Output layer with 1 neuron and sigmoid activation

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Print the model summary
model.summary()

# Assuming you have training data X_train as a DataFrame and labels y_train
X_train_array = X_train.values.reshape((X_train.shape[0], X_train.shape[1], 1))

# Train the model and store the history
history = model.fit(X_train_array, y_train, epochs=100, batch_size=32)

# Plot the loss versus accuracy graph
plt.plot(history.history['loss'], label='Loss')
plt.plot(history.history['accuracy'], label='Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Value')
plt.title('Loss and Accuracy')
plt.legend()
plt.show()

# Obtain predictions
y_pred = model.predict(X_train_array)

# Convert probabilities to class labels
y_pred_labels = np.round(y_pred).flatten()

# Generate confusion matrix
cm = confusion_matrix(y_train, y_pred_labels)

# Plot confusion matrix
plt.imshow(cm, cmap=plt.cm.Blues)
plt.title('Confusion Matrix of LSTM')
plt.colorbar()
classes = ['0 (Attack)', '1 (Natural)']  # Replace with your actual class labels
tick_marks = np.arange(len(classes))
plt.xticks(tick_marks, classes, rotation=45)
plt.yticks(tick_marks, classes)
plt.xlabel('Predicted Label',fontsize=12)
plt.ylabel('True Label',fontsize=12)

# Add labels to each cell in confusion matrix
thresh = cm.max() / 2.0
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, format(cm[i, j], 'd'), horizontalalignment="center", color="white" if cm[i, j] > thresh else "black")

plt.show()

# Generate classification report
report = classification_report(y_train, y_pred_labels, target_names=classes)
print(report)
```

## CNN

```python
X = df1.drop('marker',axis=1)
y = df1['marker']

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.1,random_state=15)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
from keras.models import Sequential
from keras.layers import Conv1D, MaxPooling1D, Flatten, Dense
plt.rcParams['font.family'] = 'Times New Roman'
# Define the model architecture
model = Sequential()
model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(128, 1)))  # Assuming input_shape is (timesteps, features)
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid'))  # Output layer with 1 neuron and sigmoid activation

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Print the model summary
model.summary()

# Assuming you have training data X_train as a DataFrame and labels y_train
X_train_array = X_train.values.reshape((X_train.shape[0], X_train.shape[1], 1))

# Train the model and store the history
history = model.fit(X_train_array, y_train, epochs=6, batch_size=16)

# Plot the loss versus accuracy graph
plt.plot(history.history['loss'], label='Loss')
plt.plot(history.history['accuracy'], label='Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Value')
plt.title('Loss and Accuracy')
plt.legend()
plt.show()

# Obtain predictions
y_pred = model.predict(X_train_array)

# Convert probabilities to class labels
y_pred_labels = np.round(y_pred).flatten()

# Generate confusion matrix
cm = confusion_matrix(y_train, y_pred_labels)

# Plot confusion matrix
plt.imshow(cm, cmap=plt.cm.Blues)
plt.title('Confusion Matrix CNN',fontsize=16)
plt.colorbar()
classes = ['0 (Attack)', '1 (Natural)']  # Replace with your actual class labels
tick_marks = np.arange(len(classes))
plt.xticks(tick_marks, classes, rotation=45)
plt.yticks(tick_marks, classes)
plt.xlabel('Predicted Label',fontsize=14)
plt.ylabel('True Label',fontsize=14)

# Add labels to each cell in confusion matrix
thresh = cm.max() / 2.0
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, format(cm[i, j], 'd'), horizontalalignment="center", color="white" if cm[i, j] > thresh else "black")

plt.show()

# Generate classification report
report = classification_report(y_train, y_pred_labels, target_names=classes)
print(report)
```

## Multi-Layer Perceptron (MLP)

```python
X = df1.drop('marker',axis=1)
y = df1['marker']

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.2,random_state=42)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
from keras.models import Sequential
from keras.layers import Dense
plt.rcParams['font.family'] = 'Times New Roman'
# Define the model architecture
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(128,)))  # Assuming input_shape is (features,)
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid'))  # Output layer with 1 neuron and sigmoid activation

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Print the model summary
model.summary()

# Assuming you have training data X_train as a DataFrame and labels y_train
X_train_array = X_train.values

# Train the model and store the history
history = model.fit(X_train_array, y_train, epochs=15, batch_size=10)

# Plot the loss versus accuracy graph
plt.plot(history.history['loss'], label='Loss')
plt.plot(history.history['accuracy'], label='Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Value')
plt.title('Loss and Accuracy')
plt.legend()
plt.show()

# Obtain predictions
y_pred = model.predict(X_train_array)

# Convert probabilities to class labels
y_pred_labels = np.round(y_pred).flatten()

# Generate confusion matrix
cm = confusion_matrix(y_train, y_pred_labels)

# Plot confusion matrix
plt.imshow(cm, cmap=plt.cm.Blues)
plt.title('Confusion Matrix MLP')
plt.colorbar()
classes = ['Attack', 'Natural']  # Replace with your actual class labels
tick_marks = np.arange(len(classes))
plt.xticks(tick_marks, classes, rotation=45)
plt.yticks(tick_marks, classes)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')

# Add labels to each cell in confusion matrix
thresh = cm.max() / 2.0
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, format(cm[i, j], 'd'), horizontalalignment="center", color="white" if cm[i, j] > thresh else "black")

plt.show()

# Generate classification report
report = classification_report(y_train, y_pred_labels, target_names=classes)
print(report)
```
