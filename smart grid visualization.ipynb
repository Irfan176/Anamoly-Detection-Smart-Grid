# Data Visualization

## Relation between 3 Phase Voltage & current Phase angles with respect to Attack or non attack for Relay R1

```python
plt.rcParams['font.family'] = 'Times New Roman'
sns.set(font_scale=1.5)
g = sns.FacetGrid(df, row='marker', height=6, aspect=2)

# Map scatter plot onto each row of the grid
g.map(sns.scatterplot, 'R1-PA1:VH', 'R1-PA4:IH', color='#FF4040', label='R1 A Phase')
g.map(sns.scatterplot, 'R1-PA2:VH', 'R1-PA5:IH', color='blue', label='R1 B Phase')
g.map(sns.scatterplot, 'R1-PA3:VH', 'R1-PA6:IH', color='green', label='R1 C Phase')

# Set titles for each row
g.set_titles(row_template='{row_name}')

# Set common x-axis label
g.set(xlabel='Voltage Phase Angle')

# Set common y-axis label
g.set(ylabel='Current Phase Angle')

# Add legend to the plot
g.add_legend(title='Phase', bbox_to_anchor=(1, 1), loc='upper left')

# Adjust the spacing between subplots
plt.tight_layout()

# Display the combined plot
plt.show()
```
